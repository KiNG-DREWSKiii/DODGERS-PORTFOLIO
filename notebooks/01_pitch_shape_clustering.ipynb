# Pitch Shape Clustering

This notebook presents a baseline workflow for grouping pitches by shape using public Statcast-style data. The objective is to show a clean, reproducible approach to an interpretable baseball research question rather than to claim finished production results.

## Question

Can public pitch-tracking features be used to identify interpretable pitch-shape clusters that may help frame pitch design and evaluation questions?

## Data

The workflow uses public Statcast-style pitch-level data through `pybaseball`. For a first pass, the analysis focuses on a narrow but widely available set of variables:

- release speed
- horizontal movement (`pfx_x`)
- vertical movement (`pfx_z`)
- plate location
- pitch type labels for interpretation only

```python
from src.project1_pitch_shape.data_pull import fetch_statcast_pitch_data
from src.project1_pitch_shape.features import build_pitch_shape_features
from src.project1_pitch_shape.clustering import run_pitch_clustering

START_DATE = "2024-03-28"
END_DATE = "2024-04-30"
N_CLUSTERS = 6
```

## Pull and prepare data

The first step is to pull a public pitch-level sample and convert movement fields into a compact pitch-shape feature table.

```python
raw = fetch_statcast_pitch_data(START_DATE, END_DATE)
features = build_pitch_shape_features(raw)
features[["pitch_type", "velo", "horizontal_break_inches", "induced_vertical_break_inches"]].head()
```

## Cluster pitch shapes

A baseline KMeans model is used after standardizing movement and velocity inputs. This is a starting point for exploratory structure, not a claim that pitch shapes are naturally spherical or fully captured by this feature set.

```python
result = run_pitch_clustering(features, n_clusters=N_CLUSTERS)
result.cluster_summary
```

## Takeaway

This baseline workflow creates a usable research starting point for questions about pitch families, pitcher-specific arsenals, and design-oriented movement bands. The next layer is to evaluate whether these clusters connect to outcomes such as whiff rate, chase rate, command profile, or run value.
