# Pitch Shape Clustering

Pitch-shape clustering analysis using public Statcast-style data.

## Question

Can pitch movement and velocity features be grouped into interpretable pitch-shape clusters?

## Data

Statcast pitch data was retrieved using `pybaseball`.

```python
from src.project1_pitch_shape.data_pull import fetch_statcast_pitch_data
from src.project1_pitch_shape.features import build_pitch_shape_features
from src.project1_pitch_shape.clustering import run_pitch_clustering

START_DATE = "2024-03-28"
END_DATE = "2024-04-30"
N_CLUSTERS = 6
```

## Feature Preparation

Movement and velocity variables were standardized prior to clustering.

```python
raw = fetch_statcast_pitch_data(START_DATE, END_DATE)
features = build_pitch_shape_features(raw)
features[["pitch_type", "velo", "horizontal_break_inches", "induced_vertical_break_inches"]].head()
```

## Clustering

KMeans clustering was applied with `k=6` clusters.

```python
result = run_pitch_clustering(features, n_clusters=N_CLUSTERS)
result.cluster_summary
```

## Notes

Cluster summaries can be compared against pitch labels, pitcher subsets, and outcome metrics in follow-up analysis.
